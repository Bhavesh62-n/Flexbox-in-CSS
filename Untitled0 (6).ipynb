```python
1. Flexbox vs Grid Layout

CSS Flexbox:

One-dimensional layout system (either row or column).

Best for aligning items along a single axis.

Use when you have a simple, linear layout.


CSS Grid:

Two-dimensional layout system (rows and columns).

Better for building complex layouts.

Use when you need full control over both axes.


When to choose:

Use Flexbox for components like navbars or toolbars.

Use Grid for full-page layouts or complex section layouts.
```

Q2. Flexbox Properties Explained

1. justify-content – Aligns items horizontally (main axis).

Values: flex-start, center, space-between, etc.



2. align-items – Aligns items vertically (cross axis).

Values: stretch, center, flex-start, etc.



3. gap – Defines the spacing between items.

Example: gap: 10px;



4. flex-direction – Sets the direction of the main axis.

Values: row, row-reverse, column, column-reverse



5. flex-wrap – Determines whether items should wrap to the next line.

Values: nowrap, wrap, wrap-reverse
